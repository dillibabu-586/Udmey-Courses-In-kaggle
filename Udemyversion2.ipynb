```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['level'] = le.fit_transform(data['level'])
data['subject'] = le.fit_transform(data['subject'])
data['course_title'] = le.fit_transform(data['course_title'])
```

```python
data.info()
```

```python
data['is_paid'] = data['is_paid'].str.replace('True','1')
data['is_paid'] = data['is_paid'].str.replace('False','0')
data['is_paid'] = data['is_paid'].str.replace('FALSE','0')
data['is_paid'] = data['is_paid'].str.replace('TRUE','1')
```

```python
data['is_paid']=data['is_paid'].astype(int)
```

```python
data.corr()
```

```python
from sklearn import linear_model, metrics 
```

```python
x=data[['course_id','course_title','is_paid','price','num_reviews','num_lectures','level','content_duration','subject']]
y=data['num_subscribers']
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4,random_state=1) 
  
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
import numpy as np
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
reg.score(x,y)
```
