```python
data.level.unique()
```

```python
data['is_paid'] = data['is_paid'].str.replace('True','1')
data['is_paid'] = data['is_paid'].str.replace('False','0')
data['is_paid'] = data['is_paid'].str.replace('FALSE','0')
data['is_paid'] = data['is_paid'].str.replace('TRUE','1')
```

```python
data.head(5)
```

```python
data.info()
```

```python
data['content_duration'] = data.content_duration.str.split(expand=True)
```

```python
data
```

```python
data.head(5)
```

```python
data['price'].unique()
data['price'] = data['price'].str.replace('Free','0')
data['price']=data['price'].astype(int)
```

```python
data['content_duration'] = data['content_duration'].astype(float)
```

```python
data['content_duration'] = data['content_duration']*60
```

```python
data['content_duration'] = data['content_duration'].astype(int)
```
